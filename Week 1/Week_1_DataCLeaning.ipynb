Same event is repeated thrice, With different categories and the location is mentioned only once.

```python
df.shape
```

```python
# Group by 'Event_name' and 'datetime', then aggregate categories and location_extracted
result = df.groupby(['Event_name', 'datetime', 'description']).agg({
    'host_org': lambda x: ', '.join(x[x != 'Unknown'].unique()),
    'event_perks': lambda x: ', '.join(x[x != 'None'].unique()),
    'categories': lambda x: ', '.join(x[x != "Not Specified"].unique()),
    'location_extracted': lambda x: ', '.join(x[x != 'Unknown'].unique())  # Exclude 'Unknown' locations
}).reset_index()
```

```python
result.shape
```

```python
# Rename columns
result.columns = ['event_name', 'datetime', 'description', 'host_org', 'event_perks', 'categories', 'location']
result.head()
```

```python
result.isnull().sum()
```

```python
result.event_perks.value_counts()
```

```python
result.replace('', np.nan, inplace=True)
result.isnull().sum()
```

```python
result.groupby('event_name')['datetime'].value_counts()[result.groupby('event_name')['datetime'].value_counts() > 1]
```
