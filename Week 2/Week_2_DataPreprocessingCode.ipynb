```python
# Group by 'Event_name' and 'datetime', then aggregate categories and location_extracted
result = df.groupby(['Event_name', 'datetime', 'description']).agg({
    'host_org': lambda x: ', '.join(x[x != 'Unknown'].unique()),
    'event_perks': lambda x: ', '.join(x[x != 'None'].unique()),
    'categories': lambda x: ', '.join(x[x != "Not Specified"].unique()),
    'location_extracted': lambda x: ', '.join(x[x != 'Unknown'].unique())  # Exclude 'Unknown' locations
}).reset_index()
```

### Week 2

```python
# Data loading in pandas Dataframe
df = pd.read_csv("/content/drive/MyDrive/AI Society Fall23/cleaned_data.csv")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
print(df.datetime.dtype)
```

```python
df.datetime = pd.to_datetime(df.datetime)
```

```python
print(df.datetime.dtype)
```

```python
# Extract features from datetime
df['event_year'] = df['datetime'].dt.year
df['event_month'] = df['datetime'].dt.month
df['event_day'] = df['datetime'].dt.day
df['event_hour'] = df['datetime'].dt.hour
df['event_dayofweek'] = df['datetime'].dt.dayofweek
df['event_weekend'] = df['datetime'].dt.dayofweek >= 5  # boolean indicating if the event is on a weekend
```

```python
df['event_weekend'] = df['event_weekend'].astype(int)
```

```python
df.head()
```

## **Data Visualization**

```python
# Distribution of events over months
plt.figure(figsize=(10, 6))
sns.countplot(x='event_month', data=df)
plt.title('Distribution of Events over Months')
plt.xlabel('Month')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Distribution of events over days of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='event_dayofweek', data=df)
plt.title('Distribution of Events over Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Events')
plt.show()
```

```python
# Most common event categories
plt.figure(figsize=(10, 6))
df['categories'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Event Categories')
plt.xlabel('Category')
plt.ylabel('Number of Events')
plt.show()
```
